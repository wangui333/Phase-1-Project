```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read the csv into a DataFrame
df = pd.read_csv('AviationData.csv', encoding='latin1')
```

**Data Cleaning**

```python
# Shape & columns
print(f"Rows × Cols: {df.shape}")
print(df.columns.tolist())
```

```python
df = df.map(lambda x: x.title() if isinstance(x, str) else x)
```

```python
#missing data
df.isna()
```

```python
df.fillna("Unknown", inplace=True)
```

```python
df.isna()
```

```python
df[df.duplicated()]
```

```python
#duplicated data
df.drop_duplicates(inplace=True)
```

```python
#remove empty rows
df=df.dropna(axis=0 , how='any')
```

```python
#filling missing values
# df=df.fillna('Unknown') # This cell is redundant and potentially problematic
```

```python
#saving data
df.to_csv('cleaned_data.csv', index=0)
```

**Imputation**

```python
#Filling all missing values
 df.fillna('unknown', inplace=True)
```

```python
#Filling missing values with mean
 df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
#Median imputation
 df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
#mode imputation
#df.fillna(df.mode().iloc[0], inplace=True)
```

```python
#Constant imputation
df.fillna(0, inplace=True)
```

**Visualization**

```python
#Bar chart for frequent accidents
ax = df['Total.Serious.Injuries'].value_counts().plot(kind='bar')
ax.set_title('Frequent Accidents')
ax.set_xlabel('Serious Injuries')
ax.set_ylabel('Count')
plt.show()
```

```python
#pie chart
df.head(5)['Total.Serious.Injuries'].value_counts().plot(kind='pie', autopct='%1.0f%%')
plt.title('Serious Injuries ')
plt.show()
```

```python
#Boxplot
df['Total.Serious.Injuries'] = pd.to_numeric(df['Total.Serious.Injuries'], errors='coerce')
df['Total.Serious.Injuries'] = df['Total.Serious.Injuries'].fillna(0)
df.boxplot(column='Total.Serious.Injuries')
plt.title('Serious Injuries')
plt.show()
```

```python
#scatter plot
df.tail(50).plot(kind='scatter', x='Injury.Severity', y='Total.Serious.Injuries')
plt.title('Serious Injuries vs Flight Hours')
plt.xlabel('Flight Hours')
plt.ylabel('Total Serious Injuries')
plt.show()
```
